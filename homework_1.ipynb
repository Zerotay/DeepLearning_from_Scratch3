```python
import numpy as np
import sklearn.datasets as dat
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
X, y = dat.make_circles(n_samples=2000, factor=0.3, noise=0.05, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)
```

```python
_, (train_ax, test_ax) = plt.subplots(ncols=2, sharex=True, sharey=True, figsize=(8, 4))

train_ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
train_ax.set_ylabel("Feature #1")
train_ax.set_xlabel("Feature #0")
train_ax.set_title("Training data")

test_ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test)
test_ax.set_xlabel("Feature #0")
_ = test_ax.set_title("Testing data")
```

$ x^2 + y ^ 2 = r ^ 2$  
이게 기본 식이고, 반지름을 1로 둬보자. 그렇다면?

```python
fig, ax = plt.subplots(1,1, figsize=(5,5))

n = 1000
x = np.sin(np.linspace(0, 2*np.pi, n))
y = np.cos(np.linspace(0, 2*np.pi, n))

ax.scatter(x,y)
plt.show()
```

아, 근데 0, 1 사이에 있는 원이어야 한다!

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
pca.fit(x.reshape(-1,1))

x_pca = pca.transform(x)
print(pca.explained_variance_ratio_)
```
